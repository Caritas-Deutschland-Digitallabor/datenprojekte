```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
df_codefor = pd.read_csv("CodeFor_Projekte.csv", sep=";")
df_codefor.head()
```

```python
url = "https://codefor.de/projekte/alle/"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
projects_soups = soup.find_all('a', class_="text-dark no-underline project-link")

project_dict = {}

for index, project in enumerate(projects_soups):
    project_dict[index] = {
        "Projektname" : project.find('h3', class_="title").text,
        "Quelle" : project.get('href'),
    }
```

```python
df_codefor["Quelle"] = [data['Quelle'] for data in project_dict.values()]
df_codefor["Projektname"] = [data['Projektname'] for data in project_dict.values()]
```

```python
df_codefor.to_csv("CodeFor_Projekte.csv")
```
